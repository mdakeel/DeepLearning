# Pytorch Training Pipeline

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import torch
```

```python
url = "https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/master/data.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
df.head()
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_test.shape,  X_train.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

```python
y_train_encoded
```

## Convert Numpy array to pytorch

```python
X_train_tensor = torch.from_numpy(X_train_scaled)
X_test_tensor = torch.from_numpy(X_test_scaled)
y_train_tensor = torch.from_numpy(y_train_encoded)
y_test_tensor = torch.from_numpy(y_test_encoded)
```

## Defining the mdoel

```python
class MySimpleNN():
  def __init__(self, x):
    self.weights = torch.rand(X.shape[1], 1, dtype= torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype = torch.float64, requires_grad=True)

  def forward(self, x):
    z= torch.matmul(x, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred
```

```python
learning_rate = 0.1
epochs = 10
```

## Create Model

```python
model = MySimpleNN(X_train_tensor)
model.weights
model.bias
```

##Defining loop

```python
for epoch in range(epochs):
    # forward pass
    y_pred = model.forward(X_train_tensor)

    # loss
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    loss = (y_train_tensor * torch.log(y_pred)) + ((1 - y_train_tensor) * torch.log(1 - y_pred))
    loss = -loss.mean()   # usually take mean and negate for binary cross-entropy

    # backward pass
    loss.backward()

    # update parameters
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad

        # zero gradients
        model.weights.grad.zero_()
        model.bias.grad.zero_()
```

##Model evaluation

```python
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  acc = (y_pred - y_test_tensor).float().mean()
  print(acc)
```
